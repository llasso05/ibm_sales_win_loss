# Preprocessing

I am using python label encoder to transform categorical values to numbers. This enables categories utilization on my algorithm.

```python
#importing module
from sklearn import preprocessing
#create object for label encoder
le = preprocessing.LabelEncoder()
#conver to categorical to numeric
df['Supplies Subgroup'] = le.fit_transform(df['Supplies Subgroup'])
df['Region'] = le.fit_transform(df['Region'])
df['Route To Market'] = le.fit_transform(df['Route To Market'])
df['Opportunity Result'] = le.fit_transform(df['Opportunity Result'])
df['Competitor Type'] = le.fit_transform(df['Competitor Type'])
df['Supplies Group'] = le.fit_transform(df['Supplies Group'])
#Display records to verify change
df.head()
```

```python
#training set
dataset = df.drop(columns=['Opportunity Number','Opportunity Result',
                           'Unnamed: 19','Unnamed: 20','Unnamed: 21'])
#setting target
target= df['Opportunity Result']
```

```python
dataset.head(n=10)
```

```python
#Gaussian sklearn module
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
#create object for gaussian
gnb = GaussianNB()
```

# Train test dataset.

In this opportunity I am picking 70% of the dataset for training and 30% for testing. Random state is set for Bias-Variance Tradeoff.

```python
#Train the ML algorithm
dataset_train, dataset_test, target_train, target_test = train_test_split(
    dataset,target, test_size = 0.30, random_state = 10)
```

```python
#create prediction from model
dataset.info()
```

```python
#creating naive bayes predictor
pred = gnb.fit(dataset_train, target_train).predict(dataset_test)
```

```python
print("Naive-Bayes accuracy : ",accuracy_score(target_test, pred, normalize = True))
```

```python
#module for linear svc
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
#create an object of type LinearSVC
svc_model = LinearSVC(random_state=0)
```

```python
#train the algorithm on training data and predict using the testing data
pred = svc_model.fit(dataset_train, target_train).predict(dataset_test)
```

```python
print("LinearSVC accuracy : ",accuracy_score(target_test, pred, normalize = True))
```

```python
#import Knearest neighbors module
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
#create object of the lassifier
neigh = KNeighborsClassifier(n_neighbors=3)
```

```python
#Train the algorithm
neigh.fit(dataset_train, target_train)
# predict the response
pred = neigh.predict(dataset_test)
```

```python
print ("KNeighbors accuracy score : ",accuracy_score(target_test, pred))
```

# Classification Report

After checking for the accuracy of all the 3 models I am using yellowbrick classifier to evaluate precision, recall and F1 for each model. 
* Precision: Evaluates acurracy of the predicted positive(true positive + False Positive)
* Recall: Evaluates true positives
* F1: is abetter measure to seek balance as it is a function between precision and Recall

```python
from yellowbrick.classifier import ClassificationReport
```

```python
# Instantiate the classification model and visualizer
visualizer = ClassificationReport(gnb, classes=['Won','Loss'])
visualizer.fit(dataset_train, target_train) 
# Fit the training data to the visualizer
visualizer.score(dataset_test, target_test) 
# Evaluate the model on the test data
g = visualizer.poof() 
# Draw/show/poof the data
```
