```python
#printing unique values to use label encoder
print("Region :",df['Region'].unique())
print("Route To Market :",df['Route To Market'].unique())
print("Opportunity Result :",df['Opportunity Result'].unique())
print("Competitor Type :",df['Competitor Type'].unique())
print("Supplies Group :",df['Supplies Group'].unique())
```

```python
#importing module
from sklearn import preprocessing
#create object for label encoder
le = preprocessing.LabelEncoder()
#conver to categorical to numeric
df['Supplies Subgroup'] = le.fit_transform(df['Supplies Subgroup'])
df['Region'] = le.fit_transform(df['Region'])
df['Route To Market'] = le.fit_transform(df['Route To Market'])
df['Opportunity Result'] = le.fit_transform(df['Opportunity Result'])
df['Competitor Type'] = le.fit_transform(df['Competitor Type'])
df['Supplies Group'] = le.fit_transform(df['Supplies Group'])
#Display records to verify change
df.head()
```

```python
#training set
dataset = df.drop(columns=['Opportunity Number','Opportunity Result','Unnamed: 19','Unnamed: 20','Unnamed: 21'])
#setting target
target= df['Opportunity Result']
```

```python
dataset.head(n=10)
```

```python
#import Knearest neighbors module
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#create object of the lassifier
neigh = KNeighborsClassifier(n_neighbors=3)
```

```python
#Train the ML algorithm
dataset_train, dataset_test, target_train, target_test = train_test_split(dataset,target, test_size = 0.30, random_state = 10)
```

```python
#Train the algorithm
neigh.fit(dataset_train, target_train)
# predict the response
pred = neigh.predict(dataset_test)
```

```python
print ("KNeighbors accuracy score : ",accuracy_score(target_test, pred))
```

```python
from yellowbrick.classifier import ClassificationReport
```

```python
visualizer = ClassificationReport(neigh, classes=['Won','Loss'])
visualizer.fit(dataset_train, target_train) 
# Fit the training data to the visualizer
visualizer.score(dataset_test, target_test) 
# Evaluate the model on the test data
g = visualizer.poof() 
# Draw/show/poof the data
```
